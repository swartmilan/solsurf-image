#### Populating of master_df
MDRL - Mean Daily Reflectance Loss, calculated as (R2-R1)/days_passed

Reflectance is an expression of the reflected average light intensity of a soiled mirror divided by the average intensity of the reflected light of that same mirror when cleaned.

```python
#### NOTE: the mean is taken of a subset of reflectance values and used as a representative reflectance value for that set,
#for the clean and soiled mirrors. The subset variability is shown for the M60 subset to give an indication of how good a 
#representation that mean reflectance value actually is. The mirrors are spaced 2m apart in all directions, therefore, the
#variability should be low.

Sets = ['S1','S2','S3','S4']
for s in Sets:
    
    ####    M30subset    ####
    #current soiled seflectance
    df_M30 = month_df.loc[(month_df['Set']==s) & (month_df['Mirror Elevation']==30)]    
    soiled_reflectance_M30 = df_M30['Soiled Reflectance'].mean()
    #reflectance loss for period -> previous clean reflectance - current soiled reflectance        PLACE SET CONDITION AFTER 1ST
    previous_reflectance_M30_df = master_df.loc[(master_df['Date']==previous_date_string) & (master_df['Set']==s)]
    previous_reflectance_M30 = previous_reflectance_M30_df.iloc[0]['Clean Reflectance M30subset']
    reflectance_loss_M30 = previous_reflectance_M30 - soiled_reflectance_M30
    #mean daily reflectance loss
    MDRL_M30 = reflectance_loss_M30 / days_passed
    #current clean reflectance
    clean_reflectance_M30 = df_M30['Clean Reflectance'].mean()
    
    ####    M60subset    ####
    #current soiled seflectance
    df_M60 = month_df.loc[(month_df['Set']==s) & (month_df['Mirror Elevation']==60)]    
    soiled_reflectance_M60 = df_M60['Soiled Reflectance'].mean()
    #reflectance loss for period -> previous clean reflectance - current soiled reflectance
    previous_reflectance_M60_df = master_df.loc[(master_df['Date']==previous_date_string) & (master_df['Set']==s)]
    previous_reflectance_M60 = previous_reflectance_M60_df.iloc[0]['Clean Reflectance M60subset']
    reflectance_loss_M60 = previous_reflectance_M60 - soiled_reflectance_M60
    #mean daily reflectance loss
    MDRL_M60 = reflectance_loss_M60 / days_passed
    #current clean reflectance
    clean_reflectance_M60 = df_M60['Clean Reflectance'].mean()
    
    
    #set variability calculation, done using only the subset of mirrors elevated at 60deg, reason being that the 30deg elevation
    #mirror are likely to soil faster and will cause the variability to be higher eventhough the comparison being made is not valid
    set_max_soiled_reflectance = df_M60['Soiled Reflectance'].max()
    set_min_soiled_reflectance = df_M60['Soiled Reflectance'].min()
    set_reflectance_range = set_max_soiled_reflectance - set_min_soiled_reflectance
    # set uncertainty and variability applies to all mirrors in set - ASSUMED
    # pull values from any mirror in sets that is at 60deg
    stats_df = month_df.loc[(month_df['Set']==s) & (month_df['Mirror']=='M3')] #any mirror will give the same value
    uncertainty = stats_df.iloc[0]['Set Measurement Uncertainty %']
    
    # DataFrame entry
    new_entry = {'Date':current_date_string,
                 'Set':s,
                 'Soiled Reflectance M30subset':soiled_reflectance_M30,
                 'Clean Reflectance M30subset':clean_reflectance_M30,
                 'MDRL M30subset':MDRL_M30,
                 'Soiled Reflectance M60subset':soiled_reflectance_M60,
                 'Clean Reflectance M60subset':clean_reflectance_M60,
                 'MDRL M60subset':MDRL_M60,
                 'Reflectance Variability M60subset':set_reflectance_range,
                 'Measurement Uncertainty %':uncertainty}
    master_df = master_df.append(new_entry, ignore_index=True)
```
